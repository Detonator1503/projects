## Bank Auction Properties
Normally bank auction process is initiated when borrower default on 3 consecutive Home Loan EMIâ€™s. A notice is served to the customer to respond within 60 days that why the bank should not initiate a auction of property due to default in payment. Borrower can pay the installments and this notice is withdrawn. Otherwise Borrower can submit his objection within 60 days with justification for non-payment of EMI. In case, no reply is received from borrower or bank is not satisfied with the reply then bank can initiate bank auction process. Once this notice expire after 60 days, bank can auction the property after 30 days.

```python
# creating a dataframe 
df = pd.DataFrame(columns = ['Listing_ID', 'Institution_Name', 'Property_Details', 'Application_Deadline',	'Auction_Date',	'Reserve_Price'])
```

```python
for row in result.tbody.find_all("tr"):
  # finding all data from the column
  column = row.find_all("td")
  
  # checking if there is the null value in the data
  if column != []:
    id = column[0].text.strip()
    i_n = column[1].text.strip()
    p_d = column[2].text.strip()
    ad = column[3].text.strip()
    aud = column[4].text.strip()
    rp = column[5].text.strip()
    df = df.append({'Listing_ID': id, 'Institution_Name': i_n, 'Property_Details': p_d, 'Application_Deadline': ad,	'Auction_Date':aud,	'Reserve_Price':rp}, ignore_index = True)

```
